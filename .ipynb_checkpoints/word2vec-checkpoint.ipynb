```python
import os
from tqdm import tqdm
import jieba
from gensim.models.word2vec import Word2Vec
```

```python
# 载入文件并合并为一个txt文件
f_txt = open('data/data.txt', 'w')
for parent, _, fileNames in os.walk('data/ChnSentiCorp_htl_unba_10000'):
    for fileName in tqdm(fileNames):
        currentPath = os.path.join(parent, fileName)
        with open(currentPath, 'r') as fp:
            f_txt.write(fp.read().decode('gbk', 'ignore').strip().encode('utf-8'))
            f_txt.write('\n')
f_txt.close()
```

```python
# 分词效果展示
with open('data/data.txt', 'r') as fp:
    str1 = fp.readline().strip()
print str1
print 
lst = jieba.lcut(str1,cut_all=False,HMM=False)
print '/'.join(lst)
```

```python
jieba.load_userdict('data/tsinghua.neg&pos.txt')
class MyText(object):
    def __iter__(self):
        for line in open('data/data.txt'):
            yield jieba.lcut(line)

sentences = MyText()
model = Word2Vec(sentences, size=100, min_count=5, workers=10)
model.save('data/w2v100_jieba')
```
